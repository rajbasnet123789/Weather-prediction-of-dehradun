```python


df.columns = df.columns.str.replace(' ', '_')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['DayOfYear'] = df['Date'].dt.dayofyear
df['WeekOfYear'] = df['Date'].dt.isocalendar().week


df = df.drop(columns=['Date'])


```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
columns_to_scale = ["Temperature", "Humidity", "DayOfYear", "WeekOfYear"]



scaler = StandardScaler()


scaled_data = scaler.fit_transform(df[columns_to_scale])


scaled_df = pd.DataFrame(scaled_data, columns=columns_to_scale)


df[columns_to_scale] = scaled_df
df
```

```python
df
```

```python
import numpy as np
import pandas as pd


df['sin_month'] = np.sin(2 * np.pi * df['Month'] / 12)
df['cos_month'] = np.cos(2 * np.pi * df['Month'] / 12)


df= df.drop(columns=['Month'])



```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Count the occurrences of each class in 'Weather_Description'
class_counts = df['Weather_Description'].value_counts()

# Identify the class with the minimum number of instances
class_to_remove = class_counts.idxmin()

# Remove the rows corresponding to the minimum class
df = df[df['Weather_Description'] != class_to_remove]

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode the 'Weather_Description' column to get numeric labels for y
df['Weather_Description_Encoded'] = label_encoder.fit_transform(df['Weather_Description'])

# Create the label mapping (string to numeric)
label_mapping = {label: i for i, label in enumerate(label_encoder.classes_)}

# Create the reverse mapping (numeric to string)
reverse_mapping = {i: label for i, label in enumerate(label_encoder.classes_)}

# Shift encoded values back to 'Weather_Description'
df['Weather_Description'] = df['Weather_Description_Encoded']

# Drop the 'Weather_Description_Encoded' column
df = df.drop(columns=['Weather_Description_Encoded'])

# Display the final DataFrame and mappings
print("DataFrame with 'Weather_Description' replaced by encoded values:")
print(df[['Weather_Description']])

print("\nLabel Mapping (String to Numeric):")
print(label_mapping)

print("\nReverse Mapping (Numeric to String):")
print(reverse_mapping)
```

```python
X=df.drop(columns='Weather_Description')
X

y=df['Weather_Description']
```

```python
y.unique()
```

```python
import numpy as np

# Assuming y is your target variable
class_counts = np.unique(y, return_counts=True)

# Display the number of instances for each class
class_counts_dict = dict(zip(class_counts[0], class_counts[1]))
print(class_counts_dict)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)


```

```python
import numpy as np

# Assuming y is your target variable
class_counts = np.unique(y_train, return_counts=True)

# Display the number of instances for each class
class_counts_dict = dict(zip(class_counts[0], class_counts[1]))
print(class_counts_dict)
```

```python
from imblearn.over_sampling import ADASYN
from imblearn.under_sampling import TomekLinks
from collections import Counter

from imblearn.combine import SMOTETomek
from collections import Counter

# Original features (X) and labels (y)
print("Original class distribution:", Counter(y))

# Define the desired class distribution
desired_class_distribution = {0: 290, 1: 280, 2: 310, 3: 300, 4: 290}

# Initialize SMOTETomek with the specified sampling strategy
smote_tomek = SMOTETomek(sampling_strategy=desired_class_distribution, random_state=42)

# Fit and resample the dataset
X_resampled, y_resampled = smote_tomek.fit_resample(X, y)

# Check the new class distribution
print("Resampled class distribution:", Counter(y_resampled))
```

```python
# Print the number of samples in the resampled data
print("Number of samples in X_resampled:", X_resampled.shape[0])

# Print the class distribution in the resampled data
unique_classes, class_counts = np.unique(y_resampled, return_counts=True)
print("Class distribution after SMOTETomek:", dict(zip(unique_classes, class_counts)))
```

```python
import numpy as np
from sklearn.utils.class_weight import compute_class_weight


# Get unique classes
classes = np.unique(y_train)

# Calculate class weights
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)

# Create a dictionary of class weights
class_weights_dict = dict(zip(classes, class_weights))

# Print the class weights
print("Class Weights:", class_weights_dict)
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import label_binarize
import numpy as np

# Adjust class weights to improve the performance of class 2
class_weights = {0: 10, 1: 9, 2: 33, 3: 8, 4: 8}

# Initialize CatBoostClassifier with updated class_weights
catboost_clf = CatBoostClassifier(
    iterations=1000,  # Keep iterations reasonable to avoid overfitting
    depth=6,         # Increased depth for better learning
    learning_rate=0.05,  # Further decreased learning rate for more stable learning
    random_state=42,
    class_weights=class_weights,
    cat_features=[],  # Specify categorical features if needed
    verbose=200       # Print progress every 200 iterations
)

# Fit the model to the training data
catboost_clf.fit(X_resampled, y_resampled)

# Predict on the test set
y_pred_catboost = catboost_clf.predict(X_test)
y_pred_proba = catboost_clf.predict_proba(X_test)

# Print classification report
print("CatBoost Classification Report:\n", classification_report(y_test, y_pred_catboost))

# Print confusion matrix
print("CatBoost Confusion Matrix:\n", confusion_matrix(y_test, y_pred_catboost))

# Binarize the output labels for ROC AUC calculation
y_test_binarized = label_binarize(y_test, classes=[0, 1, 2, 3, 4])
n_classes = y_test_binarized.shape[1]

# Calculate ROC AUC for each class
roc_auc = dict()
for i in range(n_classes):
    roc_auc[i] = roc_auc_score(y_test_binarized[:, i], y_pred_proba[:, i])

# Print ROC AUC scores for each class
for i in range(n_classes):
    print(f"ROC AUC score for class {i}: {roc_auc[i]:.2f}")

```
